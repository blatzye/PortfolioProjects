```python
#Removing the "per bottle" since all the instances use per bottle. 
#Setting the datatype as float
df["Price"] = df["Price"].str.extract(r"(\d+\.\d+)")
df["Price"] = df["Price"].astype(float)
```

```python
#Renaming the column so we know that the price is in pounds
df.rename(columns={"Price": "price_in_pounds"}, inplace=True)
```

```python
most_expensive_sort = df.groupby("Grape")["price_in_pounds"].max() 
top_10_sorts = most_expensive_sort.nlargest(10)
```

```python
plt.figure(figsize=(12, 8))
bars = plt.bar(top_10_sorts.index, top_10_sorts.values, color="navy")

for bar in bars: 
    yval = bar.get_height() 
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), 
            va="bottom", ha="center", color="black")

plt.xlabel("Sort")
plt.ylabel("Price")
plt.title("Top 10 most expensive sorts")
plt.xticks(rotation=45)
plt.show()
```
